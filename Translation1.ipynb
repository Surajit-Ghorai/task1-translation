```python
import googletrans

# checking all available language for translation
print(googletrans.LANGUAGES)
```

```python
from googletrans import Translator

translator = Translator()

def translate_text(my_text):
    translated_text = translator.translate(my_text, dest='gu', src='en')
    return translated_text
```

```python
translated_text = translate_text(text1)

print(translated_text.text)
```

```python
translated_text2 = translate_text(text2)

print(translated_text2.text)
```

```python
translated_text3 = translate_text(text3)

print(translated_text3.text)
```

# mbart-large-50-many-to-many-mmt

```python
# load model and tokenizer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
def call_mbart50(my_text):
    tokenizer.src_lang = "en_XX"  # source language = "english"
    encoded_en = tokenizer(my_text, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded_en,
        forced_bos_token_id=tokenizer.lang_code_to_id["gu_IN"]  # target language = "gujarati", for gujrati, id= "gu_IN"
    )
    return tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
translated_text = call_mbart50(text1)

print(translated_text)
```

```python
translated_text2 = call_mbart50(text2)

print(translated_text2)
```

```python
translated_text3 = call_mbart50(text3)

print(translated_text3)
```

# facebook/m2m100_418M

```python
# Load model
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/m2m100_418M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/m2m100_418M")
```

```python
def call_m2m100(my_text):
    tokenizer.src_lang = "en"  # source language = "english"
    encoded_en = tokenizer(my_text, return_tensors="pt")
    generated_tokens = model.generate(**encoded_en, forced_bos_token_id=tokenizer.get_lang_id("gu"))  # target language = "gujarati"
    return tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
translated_text = call_m2m100(text1)

print(translated_text)
```

```python
translated_text2 = call_m2m100(text2)

print(translated_text2)
```

```python
translated_text3 = call_m2m100(text3)

print(translated_text3)
```
